### Zadanie 1 [30 pkt] 
Narysuj wykresy funkcji. 
$$y=\log_{10}(x+3)-4$$
$$y=\sqrt{x-3}+2$$
Dla $ x \in [0, 2\pi]$
$$y=\sin^2 x$$ 
$$y=2\sin x$$
$$y=\sin(x+2)$$
$$y=\sin x+2$$

```python
# ùë¶=log10(ùë•+3)‚àí4
import matplotlib.pyplot as plt
import numpy as np
import math

x = np.linspace(0,2*math.pi,500)
#print(x)
y = []
for i in x:
    y.append(math.log10(i+3)-4)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("log10(x+3)-4")
plt.title("Wykres funkcji f(x) = log10(x+3)-4")
plt.show()
```

```python
# y=‚àöx-3)+2
import matplotlib.pyplot as plt
import numpy as np
import math

x = np.linspace(0,2*math.pi,500)
#print(x)
y = []
for i in x:
    y.append(math.sqrt(i+3)-4)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y=‚àöx-3)+2")
plt.title("Wykres funkcji f(x) = ‚àöx-3)+2")
plt.show()
```

```python
# ùë¶=sin^2(ùë•) czyli (1‚àícos 2x)/2

import matplotlib.pyplot as plt
import numpy as np
import math

x = np.linspace(0,2*math.pi,500)
#print(x)
y = []
for i in x:
    y.append((1-math.cos(2*i))/2)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("ùë¶=(1‚àícos 2x)/2")
plt.title("Wykres funkcji f(x) = (1‚àícos 2x)/2")
plt.show()
```

```python
# ùë¶=2sin(x)

import matplotlib.pyplot as plt
import numpy as np
import math

x = np.linspace(0,2*math.pi,500)
#print(x)
y = []
for i in x:
    y.append(2*math.sin(i))
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("ùë¶=2sin(x)")
plt.title("Wykres funkcji f(x) = 2sin(x)")
plt.show()
```

```python
#ùë¶=sin(ùë•+2)

import matplotlib.pyplot as plt
import numpy as np
import math

x = np.linspace(0,2*math.pi,500)
#print(x)
y = []
for i in x:
    y.append(math.sin(i+2))
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("ùë¶=sin(ùë•+2)")
plt.title("Wykres funkcji f(x) = sin(ùë•+2)")
plt.show()
```
