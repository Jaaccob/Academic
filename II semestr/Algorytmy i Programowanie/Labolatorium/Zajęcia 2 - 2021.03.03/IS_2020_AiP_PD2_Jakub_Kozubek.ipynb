### Zadanie 1 [30 pkt] 
Narysuj wykresy funkcji. 
$$y=\log_{10}(x+3)-4$$
$$y=\sqrt{x-3}+2$$
Dla $ x \in [0, 2\pi]$
$$y=\sin^2 x$$ 
$$y=2\sin x$$
$$y=\sin(x+2)$$
$$y=\sin x+2$$

```python
# 𝑦=log10(𝑥+3)−4
import matplotlib.pyplot as plt
import numpy as np
import math

x = np.linspace(0,2*math.pi,500)
#print(x)
y = []
for i in x:
    y.append(math.log10(i+3)-4)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("log10(x+3)-4")
plt.title("Wykres funkcji f(x) = log10(x+3)-4")
plt.show()
```

```python
# y=√x-3)+2
import matplotlib.pyplot as plt
import numpy as np
import math

x = np.linspace(0,2*math.pi,500)
#print(x)
y = []
for i in x:
    y.append(math.sqrt(i+3)-4)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y=√x-3)+2")
plt.title("Wykres funkcji f(x) = √x-3)+2")
plt.show()
```

```python
# 𝑦=sin^2(𝑥) czyli (1−cos 2x)/2

import matplotlib.pyplot as plt
import numpy as np
import math

x = np.linspace(0,2*math.pi,500)
#print(x)
y = []
for i in x:
    y.append((1-math.cos(2*i))/2)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("𝑦=(1−cos 2x)/2")
plt.title("Wykres funkcji f(x) = (1−cos 2x)/2")
plt.show()
```

```python
# 𝑦=2sin(x)

import matplotlib.pyplot as plt
import numpy as np
import math

x = np.linspace(0,2*math.pi,500)
#print(x)
y = []
for i in x:
    y.append(2*math.sin(i))
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("𝑦=2sin(x)")
plt.title("Wykres funkcji f(x) = 2sin(x)")
plt.show()
```

```python
#𝑦=sin(𝑥+2)

import matplotlib.pyplot as plt
import numpy as np
import math

x = np.linspace(0,2*math.pi,500)
#print(x)
y = []
for i in x:
    y.append(math.sin(i+2))
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("𝑦=sin(𝑥+2)")
plt.title("Wykres funkcji f(x) = sin(𝑥+2)")
plt.show()
```
